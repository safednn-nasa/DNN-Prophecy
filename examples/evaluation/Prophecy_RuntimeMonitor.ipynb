**LOAD MNIST**


```python
import numpy as np
import tensorflow as tf
import time
import keras


(x_train, y_train), (x_test_orig, y_test_orig) = keras.datasets.mnist.load_data()

len_test = int(0.5 * len(x_test_orig))
print("LEN:", len_test)

x_test = x_test_orig[0:len_test]
x_val = x_test_orig[len_test:]


y_test = y_test_orig[0:len_test]
y_val = y_test_orig[len_test:]

x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
x_val = x_val.reshape(x_val.shape[0], 28, 28, 1)

print(np.shape(x_train))
print(np.shape(x_test))
print(np.shape(x_val))

print(np.shape(y_train))
print(np.shape(y_test))
print(np.shape(y_val))



np.save('./x_train_mnist.npy', x_train)
np.save('./x_test_mnist.npy', x_test)
np.save('./x_val_mnist.npy', x_val)

np.save('./y_train_mnist.npy', y_train)
np.save('./y_test_mnist.npy', y_test)
np.save('./y_val_mnist.npy', y_val)

print('Loading the model:')
model_mnist=tf.keras.models.load_model('./dataset_models/mnist/mnist_cnn.h5')
print("Printing summary of the model:")
model_mnist.summary()
```

**LOAD CIFAR**

```python
import numpy as np

import tensorflow as tf

import time


import keras


(x_train, y_train), (x_test_orig, y_test_orig) = keras.datasets.cifar10.load_data()

len_test = int(0.5 * len(x_test_orig))
print("LEN:", len_test)

x_test = x_test_orig[0:len_test]
x_val = x_test_orig[len_test:]


y_test = y_test_orig[0:len_test]
y_val = y_test_orig[len_test:]

x_train = x_train.reshape(x_train.shape[0], 32, 32, 3)
x_test = x_test.reshape(x_test.shape[0], 32, 32, 3)
x_val = x_val.reshape(x_val.shape[0], 32, 32, 3)

print(np.shape(x_train))
print(np.shape(x_test))
print(np.shape(x_val))

print(np.shape(y_train))
print(np.shape(y_test))
print(np.shape(y_val))



np.save('./x_train_cifar.npy', x_train)
np.save('./x_test_cifar.npy', x_test)
np.save('./x_val_cifar.npy', x_val)

np.save('./y_train_cifar.npy', y_train)
np.save('./y_test_cifar.npy', y_test)
np.save('./y_val_cifar.npy', y_val)


!unzip './dataset_models/cifar/model_cifar10_conv.zip'
!pwd
!ls -lt

print('Loading the model:')
model_cifar=tf.keras.models.load_model('./model_cifar10_conv.h5')
print("Printing summary of the model:")
model_cifar.summary()
```
